```python
from imgaug import augmenters as iaa
import cv2
import os

# Путь к папке с тренировочными изображениями и масками
image_path = '/content/dataset_1.2/train/images'
mask_path = '/content/dataset_1.2/train/labels'

# Аугментатор для поворота
seq = iaa.Sequential([
    iaa.Affine(rotate=(-25, 25))  # поворот на случайный угол в диапазоне от -25 до 25 градусов
])

# Функция для аугментации изображений и масок
def augment_images(image_path, mask_path, num_augmentations=5):
    for filename in os.listdir(image_path):
        img = cv2.imread(os.path.join(image_path, filename))
        mask = cv2.imread(os.path.join(mask_path, filename), cv2.IMREAD_GRAYSCALE)

        for i in range(num_augmentations):
            images_aug = seq.augment_image(img)
            masks_aug = seq.augment_image(mask)

            cv2.imwrite(os.path.join(image_path, f"aug_{i}_{filename}"), images_aug)
            cv2.imwrite(os.path.join(mask_path, f"aug_{i}_{filename}"), masks_aug)

# Применение аугментации
augment_images(image_path, mask_path)
```

**Модель с обьединенными классами**

```python
import os
# Путь к папке с метками для тренировочного датасета
labels_train_path = '/content/dataset_1.2/train/labels/'

# Обход всех файлов в директории с метками
for filename in os.listdir(labels_train_path):
    filepath = os.path.join(labels_train_path, filename)

    # Чтение файла меток
    with open(filepath, 'r') as f:
        lines = f.readlines()

    # Обновление меток
    new_lines = []
    for line in lines:
        parts = line.strip().split()
        class_id = int(parts[0])

        # Объединение классов "Industrial"(4), "Public"(6), и "Residential"(7) в один новый класс (например, 7)
        if class_id in [4, 6, 7]:
            parts[0] = '4'

        # Обновление остальных классов, если их идентификаторы больше 7
        elif class_id > 7:
            parts[0] = str(class_id - 1)

        new_lines.append(" ".join(parts))

    # Сохранение обновленных меток
    with open(filepath, 'w') as f:
        f.write("\n".join(new_lines))
```
